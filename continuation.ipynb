```python
def clean_num_articles(value):
    if pd.isna(value):
        return pd.NA
    
    val_str = str(value).strip()
    
    # If it contains letters, it's corrupted → return NaN
    if any(char.isalpha() for char in val_str):
        return pd.NA
    
    try:
        # Remove leading zeros but preserve "0"
        cleaned = val_str.lstrip('0') or '0'
        numeric_val = float(cleaned)
        
        # Convert to int if it's a whole number
        if numeric_val.is_integer():
            numeric_val = int(numeric_val)
        
        # Articles should be non-negative
        if numeric_val >= 0:
            return numeric_val
        else:
            return pd.NA
            
    except (ValueError, TypeError):
        return pd.NA

# Apply cleaning
df["num_articles"] = df["num_articles"].apply(clean_num_articles)

# Fill NaN with forward fill or median
df["num_articles"] = df["num_articles"].fillna(method="ffill")
# OR: df["num_articles"] = df["num_articles"].fillna(df["num_articles"].median())

print("Cleaned num_articles:")
print(df["num_articles"].describe())
```

```python
def clean_goldstein_scale(value):
    if pd.isna(value):
        return pd.NA
    
    val_str = str(value)
    
    # If it contains letters, it's corrupted - return NaN
    if any(char.isalpha() for char in val_str):
        return pd.NA
    
    # Clean numeric strings
    try:
        # Remove leading zeros and convert to float
        cleaned = val_str.lstrip('0') or '0'  # Handle case where value is all zeros
        
        # Handle potential missing decimal points
        # If it's a 3-digit number like "130", it might be "1.30"
        if len(cleaned) == 3 and cleaned.isdigit():
            numeric_val = float(cleaned) / 100
        elif len(cleaned) == 2 and cleaned.isdigit() and int(cleaned) > 10:
            numeric_val = float(cleaned) / 10
        else:
            numeric_val = float(cleaned)
        
        # Goldstein scale should be between -10 and +10
        if -10 <= numeric_val <= 10:
            return numeric_val
        else:
            return pd.NA  # Outside valid range
            
    except (ValueError, TypeError):
        return pd.NA

# Apply the cleaning function
df['goldstein_scale_cleaned'] = df['goldstein_scale'].apply(clean_goldstein_scale)

# Check the results
print("Original vs cleaned comparison:")
sample_indices = df.sample(20).index
comparison = df.loc[sample_indices, ['goldstein_scale', 'goldstein_scale_cleaned']]
print(comparison)
```

```python
#Yes, you can use LabelEncoding here — it will run without errors.
#But it’s usually not a good idea because:
#LabelEncoding assigns arbitrary integers (e.g., USA=0, CHN=1, GOV=2…).
#Most ML models (like logistic regression, SVM, neural nets, k-NN) will interpret those 
#numbers as ordered/mathematical values, which can mislead the model (e.g., USA < GOV < CHN has no meaning).
#It only works “okay” with tree-based models (Random Forest, XGBoost, LightGBM, CatBoost)
#since they split on categories without assuming order.
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode the actor1_type column (overwrite the original)
le = LabelEncoder()
df['actor1_type'] = le.fit_transform(df['actor1_type'])

# Save the encoder for later use (important for making predictions on new data)
import pickle
with open('actor1_type_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)

# Check the result
print(df['actor1_type'].value_counts())
print(f"Encoded values range from 0 to {df['actor1_type'].max()}")
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Download required NLTK data (run once)
try:
    nltk.data.find('tokenizers/punkt')
    nltk.data.find('corpora/stopwords')
    nltk.data.find('corpora/wordnet')
except LookupError:
    print("Downloading NLTK data...")
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True) 
    nltk.download('wordnet', quiet=True)
    nltk.download('omw-1.4', quiet=True)
    print("NLTK data downloaded successfully!")

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    """
    Clean and preprocess text data
    """
    if pd.isna(text):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove special characters and digits, keep only letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    return text

def advanced_text_features(text):
    """
    Extract advanced text features
    """
    if pd.isna(text) or text == "":
        return {
            'word_count': 0,
            'char_count': 0,
            'sentence_count': 0,
            'avg_word_length': 0,
            'capital_ratio': 0,
            'exclamation_count': 0,
            'question_count': 0,
            'punctuation_ratio': 0
        }
    
    # Basic counts
    word_count = len(text.split())
    char_count = len(text)
    sentence_count = len(re.findall(r'[.!?]+', text))
    
    # Average word length
    words = text.split()
    avg_word_length = np.mean([len(word) for word in words]) if words else 0
    
    # Capital letters ratio
    capital_count = sum(1 for c in text if c.isupper())
    capital_ratio = capital_count / len(text) if len(text) > 0 else 0
    
    # Punctuation counts
    exclamation_count = text.count('!')
    question_count = text.count('?')
    punctuation_count = len(re.findall(r'[^\w\s]', text))
    punctuation_ratio = punctuation_count / len(text) if len(text) > 0 else 0
    
    return {
        'word_count': word_count,
        'char_count': char_count,
        'sentence_count': max(sentence_count, 1),  # At least 1
        'avg_word_length': avg_word_length,
        'capital_ratio': capital_ratio,
        'exclamation_count': exclamation_count,
        'question_count': question_count,
        'punctuation_ratio': punctuation_ratio
    }

def tokenize_and_lemmatize(text):
    """
    Tokenize and lemmatize text
    """
    if pd.isna(text) or text == "":
        return ""
    
    try:
        # Tokenize
        tokens = word_tokenize(text)
        
        # Remove stopwords and lemmatize
        tokens = [lemmatizer.lemmatize(token) for token in tokens 
                  if token not in stop_words and len(token) > 2]
        
        return ' '.join(tokens)
    except Exception as e:
        print(f"Error processing text: {str(e)[:100]}...")
        return text  # Return original text if processing fails

print("Starting text preprocessing...")

# 1. Create a copy for processing
df_processed = merged_df.copy()

# 2. Clean text data
print("Cleaning text data...")
df_processed['text_cleaned'] = df_processed['text'].apply(clean_text)

# 3. Extract text features
print("Extracting text features...")
text_features = df_processed['text'].apply(advanced_text_features)
text_features_df = pd.DataFrame(text_features.tolist())

# Add text features to dataframe
df_processed = pd.concat([df_processed, text_features_df], axis=1)

# 4. Process text for NLP
print("Tokenizing and lemmatizing...")
df_processed['text_processed'] = df_processed['text_cleaned'].apply(tokenize_and_lemmatize)

# 5. Create TF-IDF features
print("Creating TF-IDF features...")
# Use top 5000 features to avoid memory issues
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),  # Include unigrams and bigrams
    min_df=5,  # Ignore terms that appear in fewer than 5 documents
    max_df=0.95,  # Ignore terms that appear in more than 95% of documents
    stop_words='english'
)

tfidf_features = tfidf.fit_transform(df_processed['text_processed'])
tfidf_feature_names = [f'tfidf_{i}' for i in range(tfidf_features.shape[1])]

# Convert to DataFrame
tfidf_df = pd.DataFrame(tfidf_features.toarray(), columns=tfidf_feature_names)

# 6. Encode categorical variables
print("Encoding categorical variables...")

# First, let's check what types we have in actor1_country_code
print("Unique data types in actor1_country_code:")
print(df_processed['actor1_country_code'].apply(type).value_counts())

# Convert all values to strings to handle mixed types
df_processed['actor1_country_code'] = df_processed['actor1_country_code'].astype(str)

# Now encode
le_country = LabelEncoder()
df_processed['actor1_country_encoded'] = le_country.fit_transform(df_processed['actor1_country_code'])

# 7. Prepare feature matrix
print("Preparing feature matrix...")

# Select numerical features
numerical_features = [
    'actor1_code', 'actor1_type', 'actor1_code_num', 'event_code', 
    'event_root_code', 'quad_class', 'goldstein_scale', 'num_mentions', 
    'num_sources', 'num_articles', 'avg_tone', 'actor1_country_encoded'
]

# Text-based features
text_feature_cols = [
    'word_count', 'char_count', 'sentence_count', 'avg_word_length',
    'capital_ratio', 'exclamation_count', 'question_count', 'punctuation_ratio'
]

# Combine all features
X_numerical = df_processed[numerical_features + text_feature_cols]
X_tfidf = tfidf_df

# Combine numerical and text features
X_combined = pd.concat([X_numerical.reset_index(drop=True), X_tfidf.reset_index(drop=True)], axis=1)
y = df_processed['trust_factor']

print(f"Feature matrix shape: {X_combined.shape}")
print(f"Target variable shape: {y.shape}")

# 8. Display feature importance preview
print("\nText feature statistics by trust factor:")
for feature in text_feature_cols:
    print(f"\n{feature}:")
    print(df_processed.groupby('trust_factor')[feature].agg(['mean', 'std']))

# 9. Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\nTraining set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
print(f"Training set class distribution:")
print(y_train.value_counts(normalize=True))

# 10. Scale numerical features (only the first part, not TF-IDF)
print("Scaling features...")
scaler = StandardScaler()
num_features_count = len(numerical_features + text_feature_cols)

# Scale only numerical features, leave TF-IDF as is
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled.iloc[:, :num_features_count] = scaler.fit_transform(X_train.iloc[:, :num_features_count])
X_test_scaled.iloc[:, :num_features_count] = scaler.transform(X_test.iloc[:, :num_features_count])

print("Feature engineering completed!")
print(f"Ready for model training with {X_train_scaled.shape[1]} features")

# Save the TF-IDF vectorizer and scaler for later use
import pickle
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(tfidf, f)
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
with open('country_encoder.pkl', 'wb') as f:
    pickle.dump(le_country, f)

print("Preprocessing objects saved!")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import time
from sklearn.metrics import roc_curve

print("BUILDING TEXT-ONLY MODEL (No Data Leakage)")
print("="*60)

# Use only text features and TF-IDF, exclude the leaky numerical features
text_only_features = text_feature_cols  # word_count, char_count, etc.

# Create feature matrix with only text features and TF-IDF
X_text_only = pd.concat([
    df_processed[text_only_features].reset_index(drop=True),
    tfidf_df.reset_index(drop=True)
], axis=1)

print(f"Text-only feature matrix shape: {X_text_only.shape}")

# Split the data properly
X_train_text, X_test_text, y_train_text, y_test_text = train_test_split(
    X_text_only, y, 
    test_size=0.2, 
    random_state=42, 
    stratify=y,
    shuffle=True
)

# Scale only the text features (not TF-IDF)
scaler_text = StandardScaler()
num_text_features = len(text_only_features)

X_train_text_scaled = X_train_text.copy()
X_test_text_scaled = X_test_text.copy()

X_train_text_scaled.iloc[:, :num_text_features] = scaler_text.fit_transform(X_train_text.iloc[:, :num_text_features])
X_test_text_scaled.iloc[:, :num_text_features] = scaler_text.transform(X_test_text.iloc[:, :num_text_features])

# Train models on text-only features
text_models = {
    'Logistic Regression (Text Only)': LogisticRegression(random_state=42, max_iter=100),
    'Logistic Regression Balanced': LogisticRegression(random_state=42, max_iter=100, class_weight='balanced'),
    'Random Forest (Text Only)': RandomForestClassifier(n_estimators=20, random_state=42, n_jobs=-1),
    'Random Forest Balanced': RandomForestClassifier(n_estimators=20, random_state=42, n_jobs=-1, class_weight='balanced'),
#     'Gradient Boosting (Text Only)': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

text_results = {}

for name, model in text_models.items():
    print(f"\nTraining {name}...")
    
    start_time = time.time()
    model.fit(X_train_text_scaled, y_train_text)
    
    y_pred = model.predict(X_test_text_scaled)
    y_pred_proba = model.predict_proba(X_test_text_scaled)[:, 1]
    
    accuracy = model.score(X_test_text_scaled, y_test_text)
    auc_score = roc_auc_score(y_test_text, y_pred_proba)
    
    training_time = time.time() - start_time
    
    text_results[name] = {
        'model': model,
        'accuracy': accuracy,
        'auc': auc_score,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'training_time': training_time
    }
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"AUC Score: {auc_score:.4f}")
    print(f"Training time: {training_time:.2f} seconds")
    
    print("\nClassification Report:")
    print(classification_report(y_test_text, y_pred, digits=4))

# Compare text-only results
print(f"\n{'='*60}")
print("TEXT-ONLY MODEL COMPARISON")
print(f"{'='*60}")

text_comparison = pd.DataFrame({
    'Model': list(text_results.keys()),
    'Accuracy': [text_results[model]['accuracy'] for model in text_results.keys()],
    'AUC Score': [text_results[model]['auc'] for model in text_results.keys()],
    'Training Time': [text_results[model]['training_time'] for model in text_results.keys()]
})

text_comparison = text_comparison.sort_values('AUC Score', ascending=False)
print(text_comparison.to_string(index=False))

# Visualize the realistic results
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# ROC Curves
ax1 = axes[0, 0]
for name in text_results.keys():
    fpr, tpr, _ = roc_curve(y_test_text, text_results[name]['y_pred_proba'])
    auc = text_results[name]['auc']
    ax1.plot(fpr, tpr, label=f'{name.split("(")[0].strip()} (AUC = {auc:.3f})', linewidth=2)

ax1.plot([0, 1], [0, 1], 'k--', label='Random (AUC = 0.5)')
ax1.set_xlabel('False Positive Rate')
ax1.set_ylabel('True Positive Rate')
ax1.set_title('ROC Curves - Text-Only Models')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Accuracy Comparison
ax2 = axes[0, 1]
models_names = [name.replace(' (Text Only)', '').replace('Logistic Regression Balanced', 'LR Balanced').replace('Random Forest Balanced', 'RF Balanced') for name in text_results.keys()]
accuracies = [text_results[model]['accuracy'] for model in text_results.keys()]
bars = ax2.bar(range(len(models_names)), accuracies, color='lightcoral', edgecolor='darkred')
ax2.set_xlabel('Models')
ax2.set_ylabel('Accuracy')
ax2.set_title('Realistic Model Accuracy Comparison')
ax2.set_xticks(range(len(models_names)))
ax2.set_xticklabels(models_names, rotation=45, ha='right')
ax2.grid(True, alpha=0.3)

# Add value labels on bars
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.001,
             f'{acc:.3f}', ha='center', va='bottom')

# Best model confusion matrix
best_text_model_name = text_comparison.iloc[0]['Model']
best_text_predictions = text_results[best_text_model_name]['y_pred']

ax3 = axes[1, 0]
cm = confusion_matrix(y_test_text, best_text_predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', ax=ax3)
ax3.set_title(f'Confusion Matrix - {best_text_model_name.replace(" (Text Only)", "")}')
ax3.set_xlabel('Predicted')
ax3.set_ylabel('Actual')

# Feature importance for best text model
ax4 = axes[1, 1]
best_text_model_obj = text_results[best_text_model_name]['model']

if hasattr(best_text_model_obj, 'feature_importances_'):
    # Get feature names for text-only model
    text_feature_names = text_only_features + [f'tfidf_{i}' for i in range(tfidf_df.shape[1])]
    
    importances = best_text_model_obj.feature_importances_
    top_indices = np.argsort(importances)[-15:]  # Top 15 features
    
    ax4.barh(range(15), importances[top_indices], color='lightcoral')
    ax4.set_yticks(range(15))
    ax4.set_yticklabels([text_feature_names[i][:20] for i in top_indices])
    ax4.set_title(f'Top 15 Feature Importances')
    ax4.set_xlabel('Importance')
elif hasattr(best_text_model_obj, 'coef_'):
    # For logistic regression
    coefficients = abs(best_text_model_obj.coef_[0])
    text_feature_names = text_only_features + [f'tfidf_{i}' for i in range(tfidf_df.shape[1])]
    
    top_indices = np.argsort(coefficients)[-15:]
    
    ax4.barh(range(15), coefficients[top_indices], color='lightcoral')
    ax4.set_yticks(range(15))
    ax4.set_yticklabels([text_feature_names[i][:20] for i in top_indices])
    ax4.set_title(f'Top 15 Feature Coefficients')
    ax4.set_xlabel('|Coefficient|')

plt.tight_layout()
plt.show()

# Summary
best_text_model = text_comparison.iloc[0]['Model']
print(f"\nBest performing text-only model: {best_text_model}")
print(f"Realistic accuracy: {text_results[best_text_model]['accuracy']:.4f}")
print(f"AUC Score: {text_results[best_text_model]['auc']:.4f}")

# Save the best text-only model
best_model_obj = text_results[best_text_model]['model']
with open('best_text_only_trust_model.pkl', 'wb') as f:
    pickle.dump(best_model_obj, f)
    
with open('text_scaler.pkl', 'wb') as f:
    pickle.dump(scaler_text, f)

print(f"\nBest text-only model and scaler saved!")
print("Files saved: best_text_only_trust_model.pkl, text_scaler.pkl")

# Show what realistic performance looks like
if text_results[best_text_model]['accuracy'] < 1.0:
    print(f"\nThis shows realistic performance without data leakage:")
    print(f"- Accuracy: {text_results[best_text_model]['accuracy']:.1%}")
    print(f"- This is much more believable than the previous 100% accuracy")
    print(f"- The model is learning from text patterns, not perfect separators")
```
